PolynomialLinearRegression

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures

#Veri setini pandas yardımıyla dataframe df'in içine aktarıyor.
df = pd.read_csv("polynomial.csv",sep=";")

```

```python
df
```

```python
plt.scatter(df['deneyim'], df['maas'])
plt.xlabel('Deneyim(yıl)')
plt.ylabel('maas')
plt.savefig('1.png', dpi=300)
plt.show()
```

```python
#Eğer bu veri setine Linear regression uygularsak uygun olmayan bir tahmin çizgisi oluşur.

reg = LinearRegression()
reg.fit(df[['deneyim']],df['maas'])

plt.xlabel('deneyim(yıl)')
plt.ylabel('maas')

plt.scatter(df['deneyim'],df['maas'])

xekseni =df['deneyim']
yekseni= reg.predict(df[['deneyim']])
plt.plot(xekseni,yekseni,color="green", label="linear regression")
plt.legend()
plt.show()
```

```python
#Bir tane polynomial regression nesnesi oluşturmak için PolynomialFeatures fonksiyonu çağırıyoruz.
#Bu fonksiyonu çağırırken polinomun derecesni (N) belirtiyoruz.
polynomial_regression = PolynomialFeatures(degree=4)

x_polynomial = polynomial_regression.fit_transform(df[['deneyim']])
```

```python
#regression modelimizi gerçek verilerle eğitiyoruz.
reg= LinearRegression()
reg.fit(x_polynomial, df['maas'])
```

```python
y_head =reg.predict(x_polynomial)
plt.plot(df['deneyim'], y_head,color="red", label="polynomial regression")
plt.legend()

plt.scatter(df['deneyim'],df['maas'])
plt.show()
```

```python
x_polynomial1 = polynomial_regression.fit_transform([[4.5]])
reg.predict(x_polynomial1)
```
